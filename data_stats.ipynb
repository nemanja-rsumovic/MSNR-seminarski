```python
# Uzimamo relevantne kolone
df_strategies = df[['Koje strategije ste koristili prilikom učenja novog jezika?',
                    'Koliko vam je bilo zahtevno prilagoditi se novom programskom jeziku:']].copy()

# Uklanjamo prazne vrednosti
df_strategies = df_strategies.dropna()

# Split strategija na više redova
df_strategies = df_strategies.assign(
    Strategija=df_strategies['Koje strategije ste koristili prilikom učenja novog jezika?'].str.split(';')
).explode('Strategija')

# Uklanjamo vodeće/prateće razmake
df_strategies['Strategija'] = df_strategies['Strategija'].str.strip()

# Standardizujemo varijante istih strategija
df_strategies['Strategija'] = df_strategies['Strategija'].replace({
    'ChatGPT': 'ChatGPT',
    'Chatgpt': 'ChatGPT',
    'ChatGPT pomaže dosta': 'ChatGPT'
})
```

```python
# Grupisanje po strategiji i računanje proseka + broj ispitanika
strategies_summary = df_strategies.groupby('Strategija').agg(
    Prosek=('Koliko vam je bilo zahtevno prilagoditi se novom programskom jeziku:', 'mean'),
    Broj_ispitanika=('Koliko vam je bilo zahtevno prilagoditi se novom programskom jeziku:', 'count')
).sort_values('Prosek')

print(strategies_summary)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Filtriranje strategija sa najmanje 5 ispitanika
strategies_filtered = strategies_summary[strategies_summary['Broj_ispitanika'] >= 5]

# Sortiranje po proseku
strategies_filtered = strategies_filtered.sort_values('Prosek')

# Vizualizacija
plt.figure(figsize=(10,6))
sns.barplot(
    x='Prosek', 
    y=strategies_filtered.index, 
    data=strategies_filtered, 
    palette='viridis'
)

# Dodavanje broja ispitanika pored bara
for i, (prosek, broj) in enumerate(zip(strategies_filtered['Prosek'], strategies_filtered['Broj_ispitanika'])):
    plt.text(prosek + 0.05, i, f'n={broj}', va='center')

plt.xlabel('Prosečna ocena zahtevnosti ucenja novog jezika (1 = lako, 5 = teško)')
plt.ylabel('Strategija učenja')
plt.title('Efikasnost strategija učenja – prosečna zahtevnost prilagođavanja')
plt.xlim(1,5)
plt.tight_layout()
plt.show()
```

```python
# Grupisanje po frekvenciji učenja i računanje proseka + broj ispitanika
freq_summary = df.groupby('Koliko često učite nove programske jezike?').agg(
    Prosek=('Koliko vam je bilo zahtevno prilagoditi se novom programskom jeziku:', 'mean'),
    Broj_ispitanika=('Koliko vam je bilo zahtevno prilagoditi se novom programskom jeziku:', 'count')
).sort_values('Prosek')

print(freq_summary)
```
